## Inspecting a given true call

```python
def format_(df, position_):
    df = df[df['POS']==position_]    
    return pd.DataFrame(df['HG002'].iloc[0].split(':'), df['FORMAT'].iloc[0].split(':'))

position = 10616
format_(truth_set, position)
```

```python
def filter_(df, position_):    
    return df[df['POS']==position_]['FILTER'].iloc[0]

filter_(truth_set, position)
```

```python
def info_(df, position_):
    df = df[df['POS']==position_]    
    keys, values = [], []
    for field in df['INFO'].iloc[0].split(';'):
        key, value = field.split('=')
        keys.append(key)
        values.append(value)
    return pd.DataFrame(values, keys)

info_(truth_set, position)
```
